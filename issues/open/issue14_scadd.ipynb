# Add select run to an nrutils catalog file

### Setup Environment

```python
# Setup ipython environment
%load_ext autoreload
%autoreload 2
%matplotlib inline
# Import nrutils tools
from nrutils import scsearch,gwylm
from nrutils.core.nrsc import sc_add
```

```python
simdir = '/Users/book/MOUNT/tmp/mnt/hogshead/NR_data/ReducedData-hogshead/q8precessing/120_point/q8a08t150dP2_T_120_480'
database_name = 'hogshead'
# Use sc_add to append the simulation (its class representation) to the nrutils catalog 
sc_add( database_name, simdir )
```

### Look for added run in nrutils catalog using scsearch

```python
A = scsearch(keyword=simdir.split('/')[-1],verbose=True)
```

### Load Simulation

```python
y = gwylm( A[0], lmax=2, verbose=True, clean = True )
```
